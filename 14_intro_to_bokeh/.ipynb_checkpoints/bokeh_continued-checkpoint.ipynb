```python
# Basic Graph
import pandas as pd
import numpy as np

# Importing
from bokeh.plotting import figure
from bokeh.io import output_file, show
```

```python
df = pd.read_excel('verlegenhuken.xlsx')
df.head()
```

```python
# Plotting Temperature vs Pressure
x = df['Temperature']/10
y = df['Pressure']/10

# Output file
output_file('temp_air_pressure.html')

# Creating a figure object
f = figure()
f.title.text="Temperature and Air Pressure"
f.xaxis.axis_label= "Temperature (C)"
f.yaxis.axis_label= "Pressure (hPa)" 

# Line plot
f.circle(df["Temperature"],df["Pressure"],size=0.5)

# Showing the line
show(f)
```

```python
stocks = pd.read_csv('stocks.csv', parse_dates=['Date'])

p = figure(width = 800, height = 500, x_axis_type = 'datetime')

p.line(stocks['Date'], stocks['Close'], color='blue', alpha=.5)

output_file('timeseries.html')
show(p)
```
